```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.metrics.pairwise import cosine_similarity
```

```python
knowledge = {
    "Question":[
        "What is Artificial Intelligence?",
        "What is Machine Learning?",
        "What is Deep Learning?",
        "What is Python?",
        "What is Data Science?",
        "What is NLP?",
        "What is Computer Vision?",
        "What is Cloud Computing?",
        "What is Cyber Security?",
        "What is SQL?"
    ],

    "Answer":[
        "Artificial Intelligence enables machines to perform tasks that normally require human intelligence.",
        "Machine Learning is a subset of AI that learns from data.",
        "Deep Learning uses neural networks with many layers.",
        "Python is a popular programming language used in AI and data science.",
        "Data Science extracts insights from data using statistics and machine learning.",
        "Natural Language Processing helps computers understand human language.",
        "Computer Vision enables computers to interpret images and videos.",
        "Cloud Computing provides computing services over the internet.",
        "Cyber Security protects systems and networks from attacks.",
        "SQL is used to store, retrieve and manage data in databases."
    ]
}

df = pd.DataFrame(knowledge)

df
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df["Question"])
```

```python
def chatbot(user_question):

    user_vector = vectorizer.transform([user_question])

    similarity = cosine_similarity(user_vector, X)

    index = similarity.argmax()

    return df.iloc[index]["Answer"]
```

```python
question = "Explain Machine Learning"

print("User:", question)

print("Bot:", chatbot(question))
```

```python
print(chatbot("What is Python?"))

print(chatbot("Tell me about Artificial Intelligence"))

print(chatbot("Explain SQL"))

print(chatbot("What is Cyber Security?"))
```

```python
while True:

    question = input("You: ")

    if question.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    print("Chatbot:", chatbot(question))
```
